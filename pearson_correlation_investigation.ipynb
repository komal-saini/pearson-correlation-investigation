```python
import scipy.stats
import matplotlib
import matplotlib.pyplot as plt
import requests
import json

def correlation_between_two_issues(resolved_issue, suspect_issue):
    # returns pearson correlation coefficient in the form of (r, p-value)
    
    # If r = 0.4-0.9+, moderate to very strong positive association
    # If r = -0.4-(-0.9+), moderate to very strong negative association
    
    # If the p-value is low (generally less than 0.05), correlation is statistically significant
    # If the p-value is not low (generally higher than 0.05), correlation is not statistically significant 

    return scipy.stats.pearsonr(resolved_issue, suspect_issue)

time = [i for i in range(1,62)]

url = 'https://sentry.io/api/0/organizations/sentry/events-stats/'
headers = {'Authorization': '#token'}
querystring = {'interval':'1m', 'partial':1, 'project':1, 'yAxis':'count()', 'start':'2022-06-29T18:00:15','end':'2022-06-29T19:00:15'}

issue1_query = 'issue.id:2802670461'
querystring['query'] = issue1_query

issue1_data = json.loads(requests.request("GET", url, headers=headers, params=querystring).text)

issue2_query = 'issue.id:3078161811'
querystring['query'] = issue2_query

issue2_data = json.loads(requests.request("GET", url, headers=headers, params=querystring).text)

issue1_events = [events[0]["count"] for _,events in issue1_data["data"]]
issue2_events = [events[0]["count"] for _,events in issue2_data["data"]]
    
print(correlation_between_two_issues(issue1_events, issue2_events))
plt.plot(time, issue1_events, 'r')
plt.plot(time, issue2_events, 'b')
plt.show()
```
